```python
import cv2
from matplotlib import pyplot as plt

img=cv2.imread('images/0 (239).jpg',0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
height,width=img.shape
```

```python
flattenImg=img.reshape(height*width,1)
```

```python
from sklearn.cluster import KMeans

model=KMeans(n_clusters=5)
model.fit(flattenImg)
```

```python
clusteredLabels=model.labels_
```

```python
labelsToArray=clusteredLabels.reshape(height,width)
```

```python
plt.imshow(labelsToArray)
plt.colorbar()
plt.show()
```

# Note : tumor is detected as cluster number = 4

```python
import joblib

joblib.dump(model,'brain_tumor_segmentation.sav')
```
